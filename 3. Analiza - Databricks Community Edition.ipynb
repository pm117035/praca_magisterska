Wyświetlenie dostępnych plików znajdujących się na HDFS (DBFS):

```
%fs
ls /FileStore/tables
```

Załadowanie danych do ramki danych o nazwie 'd1':

```
d1 = spark.read\
       .format("csv")\
       .options(
          inferSchema="false", 
          header="true", 
          delimiter=";")\
       .load("/FileStore/tables/Dane.csv")
```

Wyświetlenie danych znajdujących się w 'd1':

```
display(d1)
```

Zaimportowanie paczki z funkcjami:

```
import pyspark.sql.functions as f
```

Pogrupowanie danych po zmiennej 'Dzien' wraz z liczbą pomiarów dla danej grupy dla ramki danych 'd1':

```
d1.groupBy("Dzien").count().display()
```

Usunięcie rekordów, w któych wartość temperatury wynosiła ponad 10°C, w celu dokładniejszej analizy:

```
d2 = d1.filter("Temperatura < 10")
```

Wyświetlenie danych znajdujących się w 'd2':

```
d2.display()
```

Pogrupowanie danych po zmiennej 'Dzien' wraz z liczbą pomiarów dla danej grupy dla ramki danych 'd2':

```
d2.groupBy("Dzien").count().display()
```

Wyświetlenie temperatury minimalnej, maksymalnej oraz średniej (zaokrąglonej do 2 miejsc po przecinku) dla pogrupowanych danych po zmiennej 'Dzien' dla ramki danych 'd2':

```
d2.groupBy('Dzien').agg(f.min('Temperatura').alias("Min T"), f.max('Temperatura').alias("MAX T"), f.round(f.avg('Temperatura'),2).alias("AVG T")).display()
```

Stworzenie tabeli tymczasowej o nazwie 'd2sql' na podstawie ramki danych 'd2':

```
d2.createOrReplaceTempView("d2sql")
```

Wypisanie dnia, minimalnej, maksymalnej oraz oraz średniej (zaokrąglonej do 2 miejsc po przecinku) dla pogrupowanych danych po zmiennej 'Dzien' dla tabeli 'd2sql' w języku SQL:

```
%sql

SELECT Dzien, MIN(Temperatura) as MIN_T, MAX(Temperatura) as MAX_T, ROUND(AVG(Temperatura),2) as AVG_T
FROM d2sql
GROUP BY Dzien
```

Pogrupowanie danych po zmiennej 'Godzina' wraz z liczbą pomiarów dla danej grupy dla ramki danych 'd2':

```
d2.groupBy("Godzina").count().display()
```

Wyświetlenie temperatury minimalnej, maksymalnej oraz średniej (zaokrąglonej do 2 miejsc po przecinku) dla pogrupowanych danych po zmiennej 'Godzina' dla ramki danych 'd2':

```
d2.groupBy("Godzina").agg(f.min("Temperatura").alias("MIN T"), f.max("Temperatura").alias("MAX T"), f.round(f.avg("Temperatura"),2).alias("AVG T")).display()
```

Wypisanie godziny, minimalnej, maksymalnej oraz oraz średniej (zaokrąglonej do 2 miejsc po przecinku) dla pogrupowanych danych po zmiennej 'Godzina' dla tabeli 'd2sql' w języku SQL:

```
%sql

SELECT Godzina, MIN(Temperatura) as MIN_T, MAX(Temperatura) as MAX_T, ROUND(AVG(Temperatura),2) as AVG_T
FROM d2sql
GROUP BY Godzina
```
